```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Download historical data for SPY
symbol = "SPY"
start_date = "2019-01-01"
end_date = "2022-12-31"

data = yf.download(symbol, start=start_date, end=end_date)

def find_big_fish(data, volume_multiplier=2):
    avg_volume = data['Volume'].mean()
    big_fish = data[data['Volume'] > avg_volume * volume_multiplier]
    return big_fish

big_fish_days = find_big_fish(data)
print(big_fish_days)
```

```python
plt.figure(figsize=(15, 7))
plt.plot(data['Close'], label='SPY Close Price', alpha=0.7)
plt.scatter(big_fish_days.index, big_fish_days['Close'], color='red', marker='o', s=100, label='Big Fish Activity')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.title('SPY Price with Big Fish Activity')
plt.show()
```
